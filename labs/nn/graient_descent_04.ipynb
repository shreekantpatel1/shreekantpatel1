# **Manual Steps**

1) Prediction \\
2) Gradients Computation: with Autograd \\
3) Loss Computation \\
4) Parameter Update

```python
import numpy as np
```

```python
# f = wx, w=2
# f = 2x

x= np.array([1, 2, 3, 4], dtype=np.float32)
y= np.array([2, 4, 6, 8], dtype=np.float32)

w = 0.0
```

```python
# Model Prediction
def forward (x):
  return w* x
```

```python
# loss 
def loss(y, y_pred):
  return ((y_pred - y)**2).mean()
```

```python
# Gradient
# MSE = 1/N * (wx - y)**2
# dJ/dw = 1/N 2x (wx - y)

def gradient(x, y, y_pred):
  return np.dot(2*x, y_pred - y).mean()
```

```python
# Before Training
forward(5)
```

```python
# Training
learning_rate = 0.01
n_iters = 100

for epoch in range(n_iters):
  # Prediction = forward Pass
  y_pred = forward(x)

  #loss
  l = loss(y, y_pred)

  # Gradients
  dw = gradient(x, y, y_pred)

  # Update weights
  w -= learning_rate * dw

  if epoch % 10 == 0:
    print(f'epoch {epoch+1}: w= {w:.3f}, loss= {l:.8f}')

```

```python
forward(5)
```

# with Torch

```python
import torch
```

```python
x= torch.tensor([1, 2, 3, 4], dtype=torch.float32)
y= torch.tensor([2, 4, 6, 8], dtype=torch.float32)

w = torch.tensor(0.0, dtype=torch.float32, requires_grad=True)
```

```python
# Before Training
forward(5)
```

```python
# Training
learning_rate = 0.01
n_iters = 100

for epoch in range(n_iters):
  # Prediction = forward Pass
  y_pred = forward(x)

  #loss
  l = loss(y, y_pred)

  # Gradients
  l.backward() # dl/dw

  # Update weights
  with torch.no_grad():
    w -= learning_rate * w.grad

  # Zero Gradient
  w.grad.zero_()

  if epoch % 10 == 0:
    print(f'epoch {epoch+1}: w= {w:.3f}, loss= {l:.8f}')
```

```python
forward(5)
```
