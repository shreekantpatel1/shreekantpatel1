# **Linear Regression**

**y_hat = wx,** \\
**loss = (y_hat - y)^2 = (wx - y)^2**

```python
x = torch.tensor(1.)
y = torch.tensor(2.)
```

```python
w = torch.tensor(1., requires_grad=True)
w
```

```python
# Forward  pass and Compute the Losss
y_hat = w * x
loss = (y_hat - y)**2
loss
```

```python
# Backward pass
loss.backward()
w.grad
```
